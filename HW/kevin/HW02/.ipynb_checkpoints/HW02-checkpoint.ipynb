### 1a

$\begin{eqnarray}
\dot{x} = 
\begin{bmatrix}
f_1(x,u) \\
f_2(x,u)
\end{bmatrix} &=& \dfrac{d}{dt}
\begin{bmatrix}
\langle i(t)\rangle \\
\langle v(t)\rangle
\end{bmatrix} \text{ where } u = 
\begin{bmatrix}
d(t) \\
\langle v_g(t)\rangle
\end{bmatrix} \\[1em]
&& \dfrac{d}{dt}
\begin{bmatrix}
\langle i(t)\rangle \\
\langle v(t)\rangle
\end{bmatrix} = 
\begin{bmatrix}
\dfrac{1}{L}\left[\langle v_g(t)\rangle-2R_{on}\langle i(t)\rangle -\langle v(t)\rangle \right]d(t) + \dfrac{1}{L}\left[-\langle v_g(t)\rangle-2R_{on}\langle i(t)\rangle -\langle v(t)\rangle \right]d'(t) \\
\dfrac{1}{C}\left[\langle i(t)\rangle-\dfrac{\langle v(t)\rangle}{R} \right]d(t)+\dfrac{1}{C}\left[\langle i(t)\rangle-\dfrac{\langle v(t)\rangle}{R} \right]d'(t)
\end{bmatrix} \\[1em]
&& \dfrac{d}{dt}
\begin{bmatrix}
\langle i(t)\rangle \\
\langle v(t)\rangle
\end{bmatrix} = 
\begin{bmatrix}
\dfrac{1}{L}\left[(2d(t)-1)\langle v_g(t)\rangle-2R_{on}\langle i(t)\rangle -\langle v(t)\rangle \right] \\
\dfrac{1}{C}\left[\langle i(t)\rangle-\dfrac{\langle v(t)\rangle}{R} \right]
\end{bmatrix} \\[1em]
&& \dfrac{d}{dt}
\begin{bmatrix}
\langle i(t)\rangle \\
\langle v(t)\rangle
\end{bmatrix} = 
\begin{bmatrix}
\dfrac{1}{L}\left[2d(t)\langle v_g(t)\rangle-\langle v_g(t)\rangle-2R_{on}\langle i(t)\rangle -\langle v(t)\rangle \right] \\
\dfrac{1}{C}\left[\langle i(t)\rangle-\dfrac{\langle v(t)\rangle}{R} \right]
\end{bmatrix}
\end{eqnarray}$

<hr>

$\begin{eqnarray}
\langle v_{L}(t)\rangle &=& L\frac{d}{dt} \left[I + \langle \hat{i}(t)\rangle \right] &=& 
\bigg[2[D+\hat{d}(t)] [V_g + \langle\hat{v_g}(t)\rangle]-[V_g + \langle \hat{v_g}(t)\rangle]-2R_{on}[I + \langle \hat{i}(t)\rangle] -[V + \langle \hat{v}(t)\rangle] \bigg]  \\[1em]
\langle v_{L}(t)\rangle &=& L\frac{d}{dt} \left[I + \langle \hat{i}(t)\rangle \right] &=& 
\bigg[2DV_g+2D\langle\hat{v_g}(t)\rangle + 2\hat{d}(t)V_g + 2\hat{d}(t)\langle\hat{v_g}(t)\rangle-V_g - \langle \hat{v_g}(t)\rangle -2R_{on}I -2R_{on}\langle \hat{i}(t)\rangle -V - \langle \hat{v}(t)\rangle \bigg]  \\[1em]
\langle \hat{v}_{L}(t)\rangle &=& L\frac{d}{dt} \langle \hat{i}(t)\rangle  &=& 
\bigg[2D\langle\hat{v_g}(t)\rangle + 2\hat{d}(t)V_g -\langle \hat{v_g}(t)\rangle -2R_{on}\langle \hat{i}(t)\rangle -\langle \hat{v}(t)\rangle \bigg]  \\[1em]
\langle \hat{v}_{L}(t)\rangle &=& L\frac{d}{dt} \langle \hat{i}(t)\rangle &=& 
\bigg[(2D-1)\langle\hat{v_g}(t)\rangle + 2\hat{d}(t)V_g -2R_{on}\langle \hat{i}(t)\rangle -\langle \hat{v}(t)\rangle \bigg]
\end{eqnarray}$

<hr>

$\begin{eqnarray}
\langle i_C (t) \rangle &=& 
C\frac{d}{dt} \left[V + \langle \hat{v}(t)\rangle \right] &=& 
\bigg[[I + \langle \hat{i}(t)\rangle]-\frac{1}{R}[V+\langle v(t)\rangle]
\bigg]  \\[1em]
\langle \hat{i}_C (t) \rangle &=& 
C\frac{d}{dt} \langle \hat{v}(t)\rangle &=& 
\bigg[\langle \hat{i}(t)\rangle -\frac{\langle v(t)\rangle}{R}
\bigg]
\end{eqnarray}$

### 1b

$\begin{eqnarray}
\frac{d}{dt} \langle \hat{i}(t)\rangle &=& \dfrac{1}{L}
\bigg[(2D-1)\langle\hat{v_g}(t)\rangle + 2\hat{d}(t)V_g -2R_{on}\langle \hat{i}(t)\rangle -\langle \hat{v}(t)\rangle \bigg] \\[1em]
\frac{d}{dt} \langle \hat{v}(t)\rangle &=& \dfrac{1}{C}
\bigg[\langle \hat{i}(t)\rangle -\frac{\langle v(t)\rangle}{R}
\bigg]
\end{eqnarray}$

<hr>

$\begin{eqnarray}
A = 
\begin{bmatrix}
\dfrac{\partial f_1}{\partial x_1} & \dfrac{\partial f_1}{\partial x_2} \\[1em] 
\dfrac{\partial f_2}{\partial x_1} & \dfrac{\partial f_2}{\partial x_2}
\end{bmatrix}\bigg|_{x,u} = 
\begin{bmatrix}
\dfrac{\partial f_1}{\partial \langle \hat{i}(t)\rangle} & \dfrac{\partial f_1}{\partial \langle \hat{v}(t)\rangle} \\[1em] 
\dfrac{\partial f_2}{\partial \langle \hat{i}(t)\rangle} & \dfrac{\partial f_2}{\partial \langle \hat{v}(t)\rangle}
\end{bmatrix}\bigg|_{x,u} = 
\begin{bmatrix}
\dfrac{-2R_{on}}{L} & \dfrac{-1}{L} \\[1em]
\dfrac{1}{C} & \dfrac{-1}{RC}
\end{bmatrix}
\end{eqnarray}$

<hr>

$\begin{eqnarray}
B = 
\begin{bmatrix}
\dfrac{\partial f_1}{\partial u_1} & \dfrac{\partial f_1}{\partial u_2} \\[1em] 
\dfrac{\partial f_2}{\partial u_1} & \dfrac{\partial f_2}{\partial u_2}
\end{bmatrix}\bigg|_{x,u} = 
\begin{bmatrix}
\dfrac{\partial f_1}{\partial \hat{d}(t)} & \dfrac{\partial f_1}{\partial \langle \hat{v_g}(t)\rangle} \\[1em] 
\dfrac{\partial f_2}{\partial \hat{d}(t)} & \dfrac{\partial f_2}{\partial \langle \hat{v_g}(t)\rangle}
\end{bmatrix}\bigg|_{x,u} = 
\begin{bmatrix}
\dfrac{-2V_g}{L} & \dfrac{2D-1}{L} \\[1em]
0 & 0
\end{bmatrix}
\end{eqnarray}$

### 1c

$\begin{eqnarray}
\hat{y}(t) &=&  \hat{x}(t) = 
\begin{bmatrix}1 & 0 \\ 0 & 1 \end{bmatrix}
\begin{bmatrix}\langle \hat{i}(t)\rangle \\ \langle \hat{v}(t)\rangle \end{bmatrix} +
\begin{bmatrix}0 & 0 \\ 0 & 0 \end{bmatrix}
\begin{bmatrix}\hat{d}(t) \\ \langle \hat{v_g}(t)\rangle \end{bmatrix} \\[1em] 
\hat{y}(s) &=&  \bigg(C(sI-A)^{-1}B +E \bigg)\hat{u}(s) \\[1em] 
G(s) &=& 
\begin{bmatrix}
G_{id}(s) & G_{ig}(s) \\[1em]
G_{vd}(s) & G_{vg}(s)
\end{bmatrix} = \\[1em]
&&\begin{bmatrix}
\dfrac{2V_g(CRs + 1)}{CLRs^2 + (2CR R_{on} + L)s + 2R_{on}+R} & \dfrac{(2D-1)(CRs+1)}{CLRs^2 + (2CR R_{on} + L)s + 2R_{on}+R} \\
\\
\dfrac{2RV_g}{CLRs^2 + (2CR R_{on} + L)s + 2R_{on}+R} & \dfrac{R(2D - 1)}{CLRs^2 + (2CR R_{on} + L)s + 2R_{on}+R}
\end{bmatrix}
\end{eqnarray}$

### 1d

```python
s,n,R_on,R,L,C,D,V_g = sp.symbols('s,n,R_on,R,L,C,D,V_g')
```

```python
A = sp.Matrix([
    [-2*R_on/L, -1/L],
    [1/C, -1/(R*C)]
])
```

```python
B = sp.Matrix([
    [2*V_g/L, (2*D-1)/L],
    [0, 0]
])
```

```python
C = sp.eye(2)
E = sp.Matrix([[0, 0],[0, 0]])
I = sp.eye(2)
```

```python
tf = C*(s*I-A).inv()*B+E
```

```python
H = sp.Matrix(tf)
```

```python
sp.simplify(H)
```

```python
sp.simplify(H[0]) 
sp.simplify(H[1]) 
sp.simplify(H[2]) 
sp.simplify(H[3]) 
```

### Duty Cycle = 0.5 with Quantization Error

```python
system = {
    'n'  :2,
    'R_on':15e-3,
    'R'  :10,
    'L'  :300e-6,
    'C'  :10e-6,
    'D'  :0.5+1.5E-5,
    'V_g' :100
}
H0 = H[0].subs(system)
H1 = H[1].subs(system)
H2 = H[2].subs(system)
H3 = H[3].subs(system)
```

```python
sp.simplify(H0)
sp.simplify(H1)
sp.simplify(H2)
sp.simplify(H3)
```

```python
f = np.logspace(0, 6, 10000)
w = 2*np.pi*f
```

```python
H0 = lambdify(s,H0,modules='numpy')
H0 = H0(1j*w)
```

```python
H1 = lambdify(s,H1,modules='numpy')
H1 = H1(1j*w)
```

```python
H2 = lambdify(s,H2,modules='numpy')
H2 = H2(1j*w)
```

```python
H3 = lambdify(s,H3,modules='numpy')
H3 = H3(1j*w)
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,10))
fsw = 25e3

ax[0].set_title('Single Phase Source Inverter Response: $d=0.50$ and $V_g=100$')
ax[0].semilogx(f,20*np.log10(abs(H0)),label='$G_{id}(s)$')
ax[0].semilogx(f,20*np.log10(abs(H1)),label='$G_{ig}(s)$')
ax[0].semilogx(f,20*np.log10(abs(H2)),label='$G_{vd}(s)$')
ax[0].semilogx(f,20*np.log10(abs(H3)),label='$G_{vg}(s)$')
#ax[0].set_xlim([1,fsw/2])
ax[0].set_ylabel('Magnitude [dB]')
ax[0].grid(which='both', axis='both')
ax[0].legend()

ax[1].semilogx(f,np.angle(H0,deg=True),label='$G_{id}(s)$')
ax[1].semilogx(f,np.angle(H1,deg=True),label='$G_{ig}(s)$',linestyle='--',dashes=(5, 5))
ax[1].semilogx(f,np.angle(H2,deg=True),label='$G_{vd}(s)$')
ax[1].semilogx(f,np.angle(H3,deg=True),label='$G_{vg}(s)$',linestyle='--',dashes=(5, 5))
#ax[1].set_xlim([1,fsw/2])
ax[1].set_ylabel('Phase [degrees]')
ax[1].set_xlabel('Frequency [Hz]')
ax[1].grid(which='both', axis='both')
ax[1].legend()
plt.show();
```

### Duty Cycle = 0.5 with NO Quantization Error

```python
s,n,R_on,R,L,C,D,V_g = sp.symbols('s,n,R_on,R,L,C,D,V_g')
```

```python
system = {
    'n'  :2,
    'R_on':15e-3,
    'R'  :10,
    'L'  :300e-6,
    'C'  :10e-6,
    'D'  :0.5,
    'V_g' :100
}
H0 = H[0].subs(system)
H1 = H[1].subs(system)
H2 = H[2].subs(system)
H3 = H[3].subs(system)
```

```python
sp.simplify(H0)
sp.simplify(H1)
sp.simplify(H2)
sp.simplify(H3)
```

```python
f = np.logspace(0, 6, 1000000)
w = 2*np.pi*f
```

```python
H0 = lambdify(s,H0,modules='numpy')
H0 = H0(1j*w)
```

```python
H1 = lambdify(s,H1,modules='numpy')
H1 = H1(1j*w)
```

```python
H2 = lambdify(s,H2,modules='numpy')
H2 = H2(1j*w)
```

```python
H3 = lambdify(s,H3,modules='numpy')
H3 = H3(1j*w)
```

```python
#Find wp (phase crossover) 
wp0 = np.where(20*np.log10(abs(H0))<=0)[0][0]
wp0label = 'wp: {:.0f} Hz'.format(f[wp0])

wp2 = np.where(20*np.log10(abs(H2))<=0)[0][0]
wp2label = 'wp:   {:.0f} Hz'.format(f[wp2])
```

```python
#Find pm (phase crossover) 
pm0label = 'pm: {:.2f} deg'.format(np.angle(H0[wp0],deg=True)+180)
pm2label = 'pm:   {:.2f} deg'.format(np.angle(H2[wp2],deg=True)+180)
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,10))
fsw = 25e3

ax[0].set_title('Single Phase Source Inverter Response: $d=0.50$ and $V_g=100$')
ax[0].semilogx(f,20*np.log10(abs(H0)),label='$G_{id}(s)$')
ax[0].semilogx(label='$G_{ig}(s)$')
ax[0].semilogx(f,20*np.log10(abs(H2)),label='$G_{vd}(s)$')
ax[0].semilogx(label='$G_{vg}(s)$')
ax[0].scatter(f[wp0],20*np.log10(abs(H0[wp0])), label=wp0label)
ax[0].scatter(f[wp2],20*np.log10(abs(H2[wp2])), label=wp2label)
#ax[0].set_xlim([1,fsw/2])
ax[0].set_ylabel('Magnitude [dB]')
ax[0].grid(which='both', axis='both')
ax[0].legend()

ax[1].semilogx(f,np.angle(H0,deg=True),label='$G_{id}(s)$')
ax[1].semilogx(label='$G_{ig}(s)$')
ax[1].semilogx(f,np.angle(H2,deg=True),label='$G_{vd}(s)$')
ax[1].semilogx(label='$G_{vg}(s)$')
ax[1].scatter(f[wp0],np.angle(H0[wp0],deg=True),label=pm0label)
ax[1].scatter(f[wp2],np.angle(H2[wp2],deg=True),label=pm2label)
#ax[1].set_xlim([1,fsw/2])
ax[1].set_ylabel('Phase [degrees]')
ax[1].set_xlabel('Frequency [Hz]')
ax[1].grid(which='both', axis='both')
ax[1].legend()
plt.show();
```
