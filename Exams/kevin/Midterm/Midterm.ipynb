### Derive capacitor voltage

```python
s,n,R_on,R,L,C,D,Dp,V_g = sp.symbols("s,n,R_on,R,L,C,D,D',V_g")
```

```python
A = sp.Matrix([
    [-R_on/L, -Dp/L],
    [Dp/C, -1/(R*C)]
])
```

```python
B = sp.Matrix([
    [(V_g/(Dp*L))*(1/(1+(R_on/(Dp**2*R)))), 1/L],
    [(-V_g/(Dp**2*R*C))*(1/(1+(R_on/(Dp**2*R)))), 0]
])
```

```python
C = sp.eye(2)
E = sp.Matrix([[0, 0],[0, 0]])
I = sp.eye(2)
tf = (s*I-A).inv()*B+E
H = sp.Matrix(tf)
```

```python
sp.simplify(H)
```

```python
C = sp.Matrix([[0, 1]])
E = sp.Matrix([[0, 0]])
I = sp.eye(2)
tf = C*(s*I-A).inv()*B+E
H = sp.Matrix(tf)
```

```python
sp.simplify(H)
```

```python
f = np.logspace(0, 7, 1000000)
w = 2*np.pi*f
tf = lambdify(s,tf,modules='numpy')
tf = tf(1j*w)
```

```python
#Find wp (gain crossover) 
wp = np.where(20*np.log10(abs(tf))<=0)[0][0]
wplabel = 'wp: {:.0f} Hz'.format(f[wp])
#Find wg (phase crossover) 
wg = np.where(np.angle(tf)>0)[0][0]
wglabel = 'wg: {:.0f} Hz'.format(f[wg])
```

```python
#Find gm (gain margin) 
gmlabel = 'gm: {:.2f} dB'.format(20*np.log10(abs(tf[wg])))
#Find pm (phase margin) 
pmlabel = 'pm: {:.2f} deg'.format(np.angle(tf[wp],deg=True)-180)
```

```python
wglabel, gmlabel
wplabel, pmlabel
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,10))

ax[0].set_title('$P(s)$')
ax[0].semilogx(f,20*np.log10(abs(tf)),label='$P(s)$')
ax[0].scatter(f[wp],20*np.log10(abs(tf[wp])), label=gmlabel,color='tab:orange',alpha=0.5)
ax[0].scatter(f[wp],20*np.log10(abs(tf[wp])), label=wglabel,color='tab:blue',alpha=0.5)
ax[0].set_ylabel('Magnitude [dB]')
ax[1].set_ylim([-150,60])
ax[0].grid(which='both', axis='both')
ax[0].legend()

ax[1].semilogx(f,np.angle(tf,deg=True),label='$P(s)$')
ax[1].scatter(f[wg],np.angle(tf[wg],deg=True),label=pmlabel,color='tab:orange',alpha=0.5)
ax[1].scatter(f[wg],np.angle(tf[wg],deg=True),label=wplabel,color='tab:blue',alpha=0.5)
ax[1].set_ylabel('Phase [degrees]')
ax[1].set_xlabel('Frequency [Hz]')
ax[1].set_ylim([-180,180])
ax[1].grid(which='both', axis='both')
ax[1].legend()
plt.show();
```
