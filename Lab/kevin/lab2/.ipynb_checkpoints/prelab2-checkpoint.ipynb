$\begin{eqnarray}
\langle v_{L}(t)\rangle &=& L\frac{d}{dt} \langle i(t)\rangle &=& 
\bigg[\bigg(\langle{v_g}(t)\rangle-\langle i(t)\rangle R_L \bigg)d(t) + \bigg( \langle{v_g}(t)\rangle-\langle i(t)\rangle R_L - \langle v(t)\rangle \bigg)d'(t) \bigg]  \\[1em]
\langle v_{L}(t)\rangle &=& L\frac{d}{dt} \langle i(t)\rangle &=& 
\bigg[\langle{v_g}(t)\rangle-\langle i(t)\rangle R_L - \langle v(t)\rangle d'(t) \bigg]  \\[1em]
V_{L}(t) &=& L\frac{d}{dt} I &=& V_g - I R_L - VD'  = 0 \\[1em]
&& VD' &=& V_g - I R_L \\[1em]
&& V &=& \dfrac{V_g - I R_L}{D'} \bigg|_{I = \dfrac{V}{D'R} }\\[1em]
&& V &=& \dfrac{V_g - \frac{V}{D'R} R_L}{D'} \\[1em]
&& V &=& \dfrac{V_g}{D'} - \dfrac{V R_L}{D'^2 R} \\[1em]
&& V + \dfrac{V R_L}{D'^2 R} &=& \dfrac{V_g}{D'} \\[1em]
&& V &=& \frac{V_g}{D'}\cdot \dfrac{1}{1+\dfrac{ R_L}{D'^2 R}} \\[1em]
\end{eqnarray}$

<hr>

$\begin{eqnarray}
&& V &=& \frac{V_g}{D'}\cdot \dfrac{1}{1+\dfrac{ R_L}{D'^2 R}} \\[1em]
&& 48 &=& \frac{24}{D'}\cdot \dfrac{1}{1+\dfrac{60\cdot10^{-3}}{D'^2 100}} \rightarrow D' = 0.4988\\[1em]
&& V &=& \frac{24}{0.4988}\cdot \dfrac{1}{1+\dfrac{ 60\cdot10^{-3}}{(0.4988)^2 100}} = 48V\\[1em]
\end{eqnarray}$

$\begin{eqnarray}
\langle i_C (t) \rangle &=& 
C\frac{d}{dt} \langle v(t)\rangle &=& 
\bigg[\bigg(\dfrac{-\langle v(t)\rangle}{R}\bigg)d(t) +\bigg(\langle i(t)\rangle -\frac{\langle v(t)\rangle}{R}d'(t) \bigg)\bigg] \\[1em]
\langle i_C (t) \rangle &=& 
C\frac{d}{dt} \langle v(t)\rangle &=& 
\bigg[\dfrac{-\langle v(t)\rangle}{R} + \langle i(t)\rangle d'(t) \bigg] \\[1em]
I_{C}(t) &=& C\frac{d}{dt} V &=& \dfrac{-V}{R} + ID' = 0  \\[1em]
&& ID' &=& \dfrac{V}{R} \\[1em]
&& I &=& \dfrac{V}{D'R} \\[1em]
&& I &=& \dfrac{1}{D'R}\cdot \frac{V_g}{D'}\cdot \dfrac{1}{1+\dfrac{ R_L}{D'^2 R}} \\[1em]
&& I &=& \frac{V_g}{D'^2 R}\cdot \dfrac{1}{1+\dfrac{ R_L}{D'^2 R}} \\[1em]
\end{eqnarray}$

$\begin{eqnarray}
\langle v_{L}(t)\rangle &=& L\frac{d}{dt} \langle i(t)\rangle &=& 
\bigg[\bigg(\langle{v_g}(t)\rangle-\langle i(t)\rangle R_L \bigg)d(t) + \bigg( \langle{v_g}(t)\rangle-\langle i(t)\rangle R_L - \langle v(t)\rangle \bigg)d'(t) \bigg]  \\[1em]
\langle v_{L}(t)\rangle &=& L\frac{d}{dt} \langle i(t)\rangle &=& 
\bigg[\langle{v_g}(t)\rangle-\langle i(t)\rangle R_L - \langle v(t)\rangle d'(t) \bigg]  \\[1em]
\langle v_{L}(t)\rangle &=& L\frac{d}{dt} \left[I + \langle \hat{i}(t)\rangle \right] &=& 
\bigg[\left[V_g + \langle\hat{v_g}(t)\rangle \right]-\left[I + \langle \hat{i}(t)\rangle \right] R_L - \left[V + \langle \hat{v}(t)\rangle \right] \left[D'-\hat{d}(t) \right] \bigg]  \\[1em]
\langle v_{L}(t)\rangle &=& L\frac{d}{dt} \left[I + \langle \hat{i}(t)\rangle \right] &=& 
\bigg[\left[V_g + \langle\hat{v_g}(t)\rangle \right]-\left[I + \langle \hat{i}(t)\rangle \right] R_L - D'V - D'\langle \hat{v}(t)\rangle + \hat{d}(t)V + \hat{d}(t)\langle \hat{v}(t)\rangle \bigg]  \\[1em]
&& \frac{d}{dt} \langle \hat{i}(t)\rangle &=& \dfrac{1}{L}
\bigg[\langle\hat{v_g}(t)\rangle -\langle \hat{i}(t)\rangle R_L - D'\langle \hat{v}(t)\rangle + \hat{d}(t)V \bigg]  \\[1em]
\end{eqnarray}$

<hr>

$\begin{eqnarray}
\langle i_C (t) \rangle &=& 
C\frac{d}{dt} \langle v(t)\rangle &=& 
\bigg[\bigg(\dfrac{-\langle v(t)\rangle}{R}\bigg)d(t) +\bigg(\langle i(t)\rangle -\frac{\langle v(t)\rangle}{R}d'(t) \bigg)\bigg] \\[1em]
\langle i_C (t) \rangle &=& 
C\frac{d}{dt} \langle v(t)\rangle &=& 
\bigg[\dfrac{-\langle v(t)\rangle}{R} + \langle i(t)\rangle d'(t) \bigg] \\[1em]
\langle i_C (t) \rangle &=& C\frac{d}{dt} \left[V + \langle \hat{v}(t)\rangle \right] &=& 
\bigg[\dfrac{-\left[V+\langle \hat{v}(t)\rangle \right]}{R} + \left[I + \langle \hat{i}(t)\rangle \right] \left[D'-\hat{d}(t) \right] \bigg] \\[1em]
\langle i_C (t) \rangle &=& C\frac{d}{dt} \left[V + \langle \hat{v}(t)\rangle \right] &=&
\bigg[\dfrac{-\left[V+\langle \hat{v}(t)\rangle \right]}{R} + ID' -I\hat{d}(t) + \langle \hat{i}(t)\rangle D' - \langle \hat{i}(t)\rangle \hat{d}(t) \bigg] \\[1em]
&& \frac{d}{dt} \langle \hat{v}(t)\rangle &=& \dfrac{1}{C}
\bigg[\dfrac{-\langle \hat{v}(t)\rangle}{R} -I\hat{d}(t) + \langle \hat{i}(t)\rangle D' \bigg] \\[1em]
\end{eqnarray}$

<hr>

$\begin{eqnarray}
\dot{x} &=& \dfrac{d}{dt}
\begin{bmatrix}
\langle \hat{i}(t)\rangle \\
\langle \hat{v}(t)\rangle
\end{bmatrix} = 
\begin{bmatrix}
\dfrac{1}{L}
\bigg[\langle\hat{v_g}(t)\rangle -\langle \hat{i}(t)\rangle R_L - D'\langle \hat{v}(t)\rangle + \hat{d}(t)V \bigg] \\
\dfrac{1}{C}
\bigg[\dfrac{-\langle \hat{v}(t)\rangle}{R} -I\hat{d}(t) + \langle \hat{i}(t)\rangle D' \bigg]
\end{bmatrix}
\end{eqnarray}$

<hr>

$\begin{eqnarray}
A = 
\begin{bmatrix}
\dfrac{\partial f_1}{\partial x_1} & \dfrac{\partial f_1}{\partial x_2} \\[1em] 
\dfrac{\partial f_2}{\partial x_1} & \dfrac{\partial f_2}{\partial x_2}
\end{bmatrix}\bigg|_{x,u} = 
\begin{bmatrix}
\dfrac{\partial f_1}{\partial \langle \hat{i}(t)\rangle} & \dfrac{\partial f_1}{\partial \langle \hat{v}(t)\rangle} \\[1em] 
\dfrac{\partial f_2}{\partial \langle \hat{i}(t)\rangle} & \dfrac{\partial f_2}{\partial \langle \hat{v}(t)\rangle}
\end{bmatrix}\bigg|_{x,u} = 
\begin{bmatrix}
\dfrac{-R_{L}}{L} & \dfrac{-D'}{L} \\[1em]
\dfrac{D'}{C} & \dfrac{-1}{RC}
\end{bmatrix}
\end{eqnarray}$

<hr>

$\begin{eqnarray}
B = 
\begin{bmatrix}
\dfrac{\partial f_1}{\partial u_1} & \dfrac{\partial f_1}{\partial u_2} \\[1em] 
\dfrac{\partial f_2}{\partial u_1} & \dfrac{\partial f_2}{\partial u_2}
\end{bmatrix}\bigg|_{x,u} = 
\begin{bmatrix}
\dfrac{\partial f_1}{\partial \hat{d}(t)} & \dfrac{\partial f_1}{\partial \langle \hat{v_g}(t)\rangle} \\[1em] 
\dfrac{\partial f_2}{\partial \hat{d}(t)} & \dfrac{\partial f_2}{\partial \langle \hat{v_g}(t)\rangle}
\end{bmatrix}\bigg|_{x,u} = 
\begin{bmatrix}
\dfrac{V}{L} & \dfrac{1}{L} \\[1em]
\dfrac{-I}{C} & 0
\end{bmatrix} = 
\begin{bmatrix}
\dfrac{V_g}{D'L}\cdot \dfrac{1}{1+\dfrac{ R_L}{D'^2 R}} & \dfrac{1}{L} \\[1em]
\dfrac{-V_g}{D'^2 RC}\cdot \dfrac{1}{1+\dfrac{ R_L}{D'^2 R}} & 0
\end{bmatrix}
\end{eqnarray}$ 

<hr>

$\begin{eqnarray}
\hat{y}(s) &=&  \bigg(C(sI-A)^{-1}B +E \bigg)\hat{u}(s) \\[1em] 
G(s) &=& 
\begin{bmatrix}
G_{id}(s) & G_{ig}(s) \\[1em]
G_{vd}(s) & G_{vg}(s)
\end{bmatrix} = \\[1em]
&&\begin{bmatrix}
\left(\dfrac{D' RV_g}{D'^2R+R_L}\right) \dfrac{RCs+2}{RLCs^2 + (RR_LC+L)s + (D'^2 R+R_L)} & 
\dfrac{RCs+1}{RLCs^2 + (RR_LC+L)s + (D'^2 R+R_L)} \\
\\
\left(\dfrac{-RV_g}{D'^2R+R_L}\right) \dfrac{Ls+(R_L-D'^2 R)}{RLCs^2 + (RR_LC+L)s + (D'^2 R+R_L)} & 
\dfrac{D'R}{RLCs^2 + (RR_LC+L)s + (D'^2 R+R_L)}
\end{bmatrix}
\end{eqnarray}$

<hr>

$\begin{eqnarray}
G_{id}(s)  = \left(\dfrac{D' RV_g}{D'^2R+R_L}\right) \dfrac{RCs+2}{RLCs^2 + (RR_LC+L)s + (D'^2 R+R_L)}
\end{eqnarray}$

<hr>

$\begin{eqnarray}
G_{vd}(s) = \left(\dfrac{-RV_g}{D'^2R+R_L}\right) \dfrac{Ls+(R_L-D'^2 R)}{RLCs^2 + (RR_LC+L)s + (D'^2 R+R_L)}
\end{eqnarray}$

<hr>

$\begin{eqnarray}
\dfrac{G_{vd}(s)}{G_{id}(s)}  = \dfrac{\left(\dfrac{-RV_g}{D'^2R+R_L}\right) \dfrac{Ls+(R_L-D'^2 R)}{RLCs^2 + (RR_LC+L)s + (D'^2 R+R_L)}}{\left(\dfrac{D' RV_g}{D'^2R+R_L}\right) \dfrac{RCs+2}{RLCs^2 + (RR_LC+L)s + (D'^2 R+R_L)}}  = \left(\dfrac{-1}{D'} \right)\dfrac{Ls+(R_L-D'^2 R)}{RCs+2}\\[2em]
\end{eqnarray}$

<hr>

```python
s,n,R_L,R,L,C,D,Dp,V_g = sp.symbols("s,n,R_L,R,L,C,D,D',V_g")
```

```python
A = sp.Matrix([
    [-R_L/L, -Dp/L],
    [Dp/C, -1/(R*C)]
])
```

```python
B = sp.Matrix([
    [(V_g/(Dp*L))*(1/(1+(R_L/(Dp**2*R)))), 1/L],
    [(-V_g/(Dp**2*R*C))*(1/(1+(R_L/(Dp**2*R)))), 0]
])
```

```python
B
```

```python
C = sp.eye(2)
E = sp.Matrix([[0, 0],[0, 0]])
I = sp.eye(2)
```

```python
tf = C*(s*I-A).inv()*B+E
```

```python
H = sp.Matrix(tf)
```

```python
sp.simplify(H)
```

```python
sp.simplify(H[0]) 
sp.simplify(H[1]) 
sp.simplify(H[2]) 
sp.simplify(H[3])
```

```python
sp.simplify(H[2]/H[0]) 
```

```python
system = {
    'R_L':60e-3,
    'R'  :100,
    'L'  :0.2e-3,
    'C'  :2200e-6,
    "D'"  :0.4988,
    'V_g' :24
}
H0 = H[0].subs(system)
H1 = H[1].subs(system)
H2 = H[2].subs(system)
H3 = H[3].subs(system)
```

```python
sp.simplify(H0)
sp.simplify(H1)
sp.simplify(H2)
sp.simplify(H3)
```

```python
sp.simplify((H[2]/H[0]).subs(system))
```

### Gid

```python
H = sp.Matrix(tf)
```

```python
f = np.logspace(0, 5, 10000)
w = 2*np.pi*f
```

```python
H0 = lambdify(s,H0,modules='numpy')
H0 = H0(1j*w)
```

```python
H0_zero = lambdify(s,s+9.09,modules='numpy')
H0_zero = H0_zero(1j*w)
```

```python
H0_pole0 = lambdify(s,1/(s+(152.27-737.31j)),modules='numpy')
H0_pole0 = H0_pole0(1j*w)
```

```python
H0_pole1 = lambdify(s,1/(s+(152.27+737.31j)),modules='numpy')
H0_pole1 = H0_pole1(1j*w)
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,10))
fsw = 25e3

ax[0].set_title('Boost Converter Response: $G_{id}(s)$')
ax[0].semilogx(f,20*np.log10(abs(H0)),label='$G_{id}(s)$')
ax[0].semilogx(f,20*np.log10(3.8492)*np.ones(len(f)),label='Gain: 4')
ax[0].semilogx(f,20*np.log10(abs(H0_zero)),label='Zero: s+9')
ax[0].semilogx(f,20*np.log10(abs(H0_pole0)),label='Pole: s+(152-737j)')
ax[0].semilogx(f,20*np.log10(abs(H0_pole1)),label='Pole: s+(152+737j)')
#ax[0].set_xlim([1,fsw/2])
ax[0].set_ylabel('Magnitude [dB]')
ax[0].grid(which='both', axis='both')
ax[0].legend()

ax[1].semilogx(f,np.angle(H0,deg=True),label='$G_{id}(s)$')
ax[1].semilogx(f,np.angle(3.8492*np.ones(len(f)),deg=True),label='Gain: 4')
ax[1].semilogx(f,np.angle(H0_zero,deg=True),label='Zero: s+9',alpha=0.75)
ax[1].semilogx(f,np.angle(H0_pole0,deg=True),label='Pole: s+(152-737j)',alpha=0.75)
ax[1].semilogx(f,np.angle(H0_pole1,deg=True),label='Pole: s+(152+737j)')
#ax[1].set_xlim([1,fsw/2])
ax[1].set_ylabel('Phase [degrees]')
ax[1].set_xlabel('Frequency [Hz]')
ax[1].grid(which='both', axis='both')
ax[1].legend()
plt.show();
```

```python
H4 = (Dp**2*R-L*s-R_L)/(Dp*(R*C*s+2))
```

```python
system = {
    'R_L':60e-3,
    'R'  :100,
    'L'  :0.2e-3,
    'C'  :2200e-6,
    "D'"  :0.4988,
    'V_g' :24
}
H4 = H4.subs(system)
```
